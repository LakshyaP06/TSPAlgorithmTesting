```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from pyCombinatorial.utils import graphs, util
from scipy.spatial.distance import cdist
import os
```

```python
def generate_cities(n, seed=42):
    np.random.seed(seed)
    return np.random.uniform(0, 500, size=(n, 2))

def display_distance_matrix(matrix):
    df = pd.DataFrame(matrix)
    print(df.round(2))
    
def create_distance_matrix(cities):
    return cdist(cities, cities, metric='euclidean')
```

```python
cities = generate_cities(20)
dist_matrix = create_distance_matrix(cities)
pd.DataFrame(dist_matrix).round(2)
```

```python
!pip install pulp

import pulp
from itertools import combinations

def tsp_ilp(distance_matrix, verbose=True):
    """
    Solves TSP using Integer Linear Programming (MTZ formulation).
    
    Args:
        distance_matrix (np.ndarray): Square distance matrix
        verbose (bool): Print solver progress
    
    Returns:
        tuple: (route, distance, optimal_flag)
    """
    n = len(distance_matrix)
    cities = range(n)
    
    prob = pulp.LpProblem("TSP_ILP", pulp.LpMinimize)
    
    x = pulp.LpVariable.dicts("x", [(i,j) for i in cities for j in cities if i != j], 
                            cat='Binary')
    
    u = pulp.LpVariable.dicts("u", cities, lowBound=0, upBound=n-1, cat='Continuous')
    
    prob += pulp.lpSum(distance_matrix[i][j] * x[(i,j)] 
                      for i in cities for j in cities if i != j)
    
    for j in cities:
        prob += pulp.lpSum(x[(i,j)] for i in cities if i != j) == 1
    
    for i in cities:
        prob += pulp.lpSum(x[(i,j)] for j in cities if i != j) == 1
    
    for i in cities:
        for j in cities:
            if i != j and (i != 0 and j != 0):
                prob += u[i] - u[j] + n*x[(i,j)] <= n - 1
    
    solver = pulp.PULP_CBC_CMD(msg=verbose)
    prob.solve(solver)
    
    status = pulp.LpStatus[prob.status]
    if status == 'Optimal':
        route = [0] 
        while len(route) < n:
            current = route[-1]
            next_city = [j for j in cities 
                        if j != current and pulp.value(x[(current,j)]) == 1][0]
            route.append(next_city)
        
        distance = sum(distance_matrix[route[i]][route[i+1]] 
                   for i in range(n-1))
        distance += distance_matrix[route[-1]][route[0]]
        
        return route, distance, True
    else:
        raise ValueError(f"Solver status: {status} - No optimal solution found")
```

```python
# path
ilp_results_file = "ilp_results.csv"
# number of samples ran
num_samples = 1
# city being tested
ilp_sizes = [4] 
# seed
fixed_seed = 42 

if os.path.exists(ilp_results_file):
    ilp_df = pd.read_csv(ilp_results_file)
    # making sure that the samples column exists
    if 'samples' not in ilp_df.columns:
        ilp_df['samples'] = 1
    print("Resuming from saved file...")
else:
    ilp_df = pd.DataFrame(columns=['cities', 'distance', 'runtime_ms', 'samples'])
    print("Starting new benchmark...")

for size in ilp_sizes:
    print(f"\nProcessing {size} cities ({num_samples} samples)...")
    
    runtimes = []
    distances = []
    
    for sample in range(num_samples):
        try:
            cities = generate_cities(size, seed=fixed_seed)
            dist_matrix = create_distance_matrix(cities)
            
            start = time.time()
            route, distance, optimal = tsp_ilp(dist_matrix, verbose=False)
            runtime = (time.time() - start) * 1000  # ms
            
            runtimes.append(runtime)
            distances.append(distance)
            
            print(f"Sample {sample+1}: {runtime:.2f} ms | Distance: {distance:.2f}")
            
        except Exception as e:
            print(f"Error in sample {sample+1}: {e}")
            continue
    
    if runtimes:
        avg_runtime = round(np.mean(runtimes), 2)
        avg_distance = round(np.mean(distances), 2)
        
        existing = ilp_df[ilp_df['cities'] == size]
        if not existing.empty:
            total_samples = existing['samples'].values[0] + len(runtimes)
            ilp_df.loc[ilp_df['cities'] == size, ['runtime_ms', 'distance', 'samples']] = [
                round((existing['runtime_ms'].values[0] * existing['samples'].values[0] + 
                      avg_runtime * len(runtimes)) / total_samples, 2),
                round((existing['distance'].values[0] * existing['samples'].values[0] + 
                      avg_distance * len(runtimes)) / total_samples, 2),
                total_samples
            ]
        else:
            # New entry
            ilp_df = pd.concat([ilp_df, pd.DataFrame([{
                'cities': size,
                'distance': avg_distance,
                'runtime_ms': avg_runtime,
                'samples': len(runtimes)
            }])], ignore_index=True)
        
        # Save results
        ilp_df.to_csv(ilp_results_file, index=False, float_format='%.2f')

print("\nILP Completed")
print(ilp_df[['cities', 'distance', 'runtime_ms', 'samples']])
```

```python
# File path
ilp_results_file = "ilp_results.csv"

if os.path.exists(ilp_results_file):
    df = pd.read_csv(ilp_results_file)

    # filter values
    df = df.sort_values(by='cities')
    df = df[df['runtime_ms'] > 0]

    # trendline
    log_cities = np.log(df['cities'])
    log_runtime = np.log(df['runtime_ms'])

    coeffs = np.polyfit(log_cities, log_runtime, 1)
    predicted_log_runtime = np.polyval(coeffs, log_cities)
    residuals = log_runtime - predicted_log_runtime

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

    # Scatter plot (Left)
    ax1.plot(df['cities'], df['runtime_ms'],
             marker='o', linestyle='--',
             color='cyan', label='ILP Runtime')
    ax1.set_title('ILP Runtime (Linear Scale)')
    ax1.set_xlabel('Number of Cities')
    ax1.set_ylabel('Runtime (ms)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # Log graph plot with trendline (middle)
    trend = np.exp(coeffs[1]) * (df['cities'] ** coeffs[0])
    ax2.loglog(df['cities'], df['runtime_ms'],
               marker='o', linestyle='--',
               color='teal', label='Runtime')
    ax2.loglog(df['cities'], trend,
               'r--', label=f'Trend (slope={coeffs[0]:.2f})')
    ax2.set_title('ILP Runtime Scaling (Log-Log)')
    ax2.set_xlabel('Number of Cities (log scale)')
    ax2.set_ylabel('Runtime (ms) (log scale)')
    ax2.grid(True, which='both', alpha=0.3)
    ax2.legend()

    # Histogram of residuals (right)
    ax3.hist(residuals, bins=10, color='cyan', edgecolor='black', alpha=0.7)
    ax3.set_title('Histogram of Log-Residuals')
    ax3.set_xlabel('Log-Residual (log(actual) - log(predicted))')
    ax3.set_ylabel('Frequency')
    ax3.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print(f"Estimated complexity exponent: {coeffs[0]:.2f}")
    if coeffs[0] > 4.5:
        print("This suggests exponential or worse-than-polynomial scaling")
    elif coeffs[0] > 3.0:
        print("This suggests cubic or quartic complexity (O(nn))")
    elif coeffs[0] > 2.0:
        print("This suggests quadratic to cubic scaling")
    else:
        print("This suggests relatively efficient scaling (sub-quadratic)")

else:
    print("Not found.")
```
